# **Promedio de Area por Barrio**

En la gráfica se muestra el top 10 de barrios con promedios de areas ofrecidas para la venta, **Bellavista Occidental** en el decimo puesto y **Modelia** de primer lugar. 

```python
agrupado=inmuebles.groupby('Barrio')['Area'].mean().reset_index().sort_values(by='Area',ascending=False)
top10=agrupado.head(10)
top10.set_index('Barrio', inplace=True)
top10.plot(kind = 'barh',title='Promedio de areas por barrios')
```
